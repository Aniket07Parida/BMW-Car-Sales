```python
import pandas as pd
import sqlite3
```

```python
# Creating a database connection
conn = sqlite3.connect("inventory.db")
```

```python
# check the table present in your database
table = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table'",conn)
table
```

```python
for tables in table['name']:
    print('-'* 50,tables,'-'* 50)
    print("Count of the records",pd.read_sql(f"select count(*) as count from `{tables}`",conn)['count'].values[0])
    display(pd.read_sql(f"select * from `{tables}`",conn))
```
